### Streaming and Batch

### streaming

```python
from dotenv import load_dotenv
import os

load_dotenv(override=True)

print(os.getenv("GROQ_API_KEY"))
```

```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

load_dotenv(override=True)

model = init_chat_model("groq:llama-3.1-8b-instant")

response = model.invoke("Hello")
print(response.content)
```

```python
for chunk in model.stream("write an essay on AI in 500 words"):
    print(chunk.content, end="", flush=True)
```

```python
model.invoke("why do parrots have colourful feathers?")
```

```python
for chunk in model.stream("why do parrots have colourful feathers?"):
    print(chunk.text, end="", flush=True)
```

### Batch 
Batching is the collection of independent request to a model can significant improve performance and reduce the cost as the processing can be done in parallel.

```python
responses = model.batch(
    [
        "what is AI?",
        "what is ML?",
        "difference between AI and ML",
    ]
)

for response in responses:
    print(response.content)
```
